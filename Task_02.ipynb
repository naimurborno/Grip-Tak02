```python
#importing libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
hours=pd.DataFrame([2.5,5.1,3.2,8.5,3.5,1.5,9.2,5.5,8.3,2.7,7.7,5.9,4.5,3.3,1.1,8.9,2.5,1.9,6.1,7.4,2.7,4.8,3.8,6.9,7.8])
scores=pd.DataFrame([21,47,27,75,30,20,88,60,81,25,85,62,41,42,17,95,30,24,67,69,30,54,35,76,86])
df=pd.concat([hours,scores],axis=1)
df.columns={'hours','scores'}
print(df)
```

```python
df.info()
```

```python
#Distribution of the data
sns.histplot(data=df,x='hours')
plt.title('Distribution of the feature(hours)')
plt.show()
sns.histplot(data=df,x='scores')
plt.title('Distribution of the target(scores)')
plt.show()
```

```python
#Dividing the data in to feature and target
X=df.drop(columns=['scores'])
y=df['scores']
```

```python
#Building the model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#Training the model
model.fit(X,y)
```

```python
sns.regplot(data=df,x='hours',y='scores',color='red',scatter_kws={'color':'black'},line_kws={'color':'red'})
```

### Question:what will be the predicted score if the student studies for 9.25hr/day?

```python
sc=np.array([9.25])
sc=sc.reshape(1,-1)
print("If the student studies for 9.25hr/day the predicted score will be",model.predict(sc)[0])
```
